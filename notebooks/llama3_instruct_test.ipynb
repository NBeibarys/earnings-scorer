# Llama 3 70B Instruct Test

This notebook checks whether we can import and interact with a local Llama 3 70B Instruct model.

## 1. Import Required Libraries

Import the packages needed to load the model and tokenizer. If these packages are not installed, install them first.

```python
import sys
import os

try:
    import torch
    from transformers import AutoTokenizer, AutoModelForCausalLM
    print('Imports successful')
except Exception as e:
    print('Import failed:', e)
```

## 2. Load the Llama 3 70B Instruct Model

Attempt to load the Llama 3 70B Instruct model from a local or online Hugging Face repository. Adjust the model name or local path if needed.

```python
model_name = 'meta-llama/Llama-3-70b-instruct'  # update if you have a different local path

hf_token = os.getenv('HUGGINGFACE_TOKEN', None)
if hf_token:
    print('Using Hugging Face token from HUGGINGFACE_TOKEN')

try:
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True, use_auth_token=hf_token)
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        device_map='auto',
        torch_dtype=torch.float16,
        use_auth_token=hf_token,
    )
    print(f'Model loaded: {model_name}')
except Exception as e:
    print('Model load failed:', e)
    if hf_token is None:
        print('If this is a private repo, set HUGGINGFACE_TOKEN or run `huggingface-cli login`.')
```

## 3. Set Up Interaction Parameters

Define the tokenizer and generation settings for a simple test prompt.

```python
prompt = 'You are a finance expert. Summarize the investment outlook from the following text: The company expects to invest in growth opportunities next year.'
gen_kwargs = {
    'max_new_tokens': 128,
    'temperature': 0.7,
    'top_p': 0.9,
    'do_sample': True,
}
print('Prompt and generation settings configured')
```

## 4. Test Basic Interaction

Run a simple prompt through the model and print the result.

```python
try:
    inputs = tokenizer(prompt, return_tensors='pt').to(model.device)
    output = model.generate(**inputs, **gen_kwargs)
    decoded = tokenizer.decode(output[0], skip_special_tokens=True)
    print('Model output:')
    print(decoded)
except NameError:
    print('Model or tokenizer not available. Load the model first.')
except Exception as e:
    print('Generation failed:', e)
```
